**Q1C) Univariate Analysis (distribution plots of all the continuous variable(s) barplots/countplots of all the categorical variables)**

**Q2) Hypothesis Testing (30 Points):**

**Q2A) 2- Sample T-Test to check if Working Day has an effect on the number of electric cycles rented (10 points)**

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('yulu.csv')

working_day = df[df['workingday'] == 1]['count']
non_working_day = df[df['workingday'] == 0]['count']

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(working_day, non_working_day)

alpha = 0.05

print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Null hypothesis rejected: Working day has a significant effect on the number of electric cycles rented.")
else:
    print("Null hypothesis not rejected: There is no significant effect of working day on the number of electric cycles rented.")
```

**Q2B) ANOVA to check if No. of cycles rented is similar or different in different 1. weather 2. season (10 points)**

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('yulu.csv')

# Perform ANOVA for 'weather' vs. 'count'
weather_groups = [df[df['weather'] == i]['count'] for i in df['weather'].unique()]
f_statistic_weather, p_value_weather = stats.f_oneway(*weather_groups)

# Significance level (alpha)
alpha = 0.05

# p-values in the output
print(f"P-Value for Weather vs. Count: {p_value_weather:.4f}")

if p_value_weather < alpha:
    print("Null hypothesis rejected: There is a significant difference in the number of cycles rented among different weather conditions.")
else:
    print("Null hypothesis not rejected: There is no significant difference in the number of cycles rented among different weather conditions.")
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('yulu.csv')

# ANOVA for 'season' vs. 'count'
season_groups = [df[df['season'] == i]['count'] for i in df['season'].unique()]
f_statistic_season, p_value_season = stats.f_oneway(*season_groups)

# Significance level (alpha)
alpha = 0.05

print(f"P-Value for Season vs. Count: {p_value_season:.4f}")

# Check the p-values against the significance level
if p_value_season < alpha:
    print("Null hypothesis rejected: There is a significant difference in the number of cycles rented among different seasons.")
else:
    print("Null hypothesis not rejected: There is no significant difference in the number of cycles rented among different seasons.")
```

**Q2C) Chi-square test to check if Weather is dependent on the season (10 points)**

```python
import pandas as pd
from scipy.stats import chi2_contingency

df = pd.read_csv('yulu.csv')

contingency_table = pd.crosstab(df['weather'], df['season'])

# Perform the Chi-square test of independence
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Define the significance level (alpha)
alpha = 0.05

print(f"P-Value: {p_value:.4f}")

# Check the p-value against the significance level
if p_value < alpha:
    print("Null hypothesis rejected: Weather is dependent on season.")
else:
    print("Null hypothesis not rejected: Weather is independent of season.")
```
